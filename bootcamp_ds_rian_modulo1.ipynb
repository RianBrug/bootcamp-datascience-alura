#AULA02 - Primeiras visualizações de dados

```
import pandas as pd

##importando dados por ano/mes processamento

dados = pd.read_csv("/content/procedimento-hospitalares-sus-por-local-internacao-valor-total-por-ano-mes-processamento.csv", encoding="ISO-8859-1",
                    skiprows = 3, sep=";", skipfooter=12,
                    thousands=".", decimal=",")
dados
```

```
dados[["2008/Ago", "2008/Set"]].head()
##fatiamento dos dados do nosso dataframe
## para realizar uma análise, necessário limpar os arquivos
```

```
colunas_usaveis = dados.mean().index.tolist()
## mean ignora as colunas vazias. desta forma, os resultados, possuem todos os valores
## eliminação das colunas vazias é de acordo com a estratégia que esta sendo usada. 
## em outras situações, poderia ser aplicado estratégias para lidar com esses campos vazios
colunas_usaveis.insert(0, "Unidade da Federação")
##inserindo o valor 'UF', necessario transformar o index (série imutável) para uma lista de python
colunas_usaveis
```

```
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```
usaveis = usaveis.set_index("Unidade da Federação")
```

```
usaveis["2019/Ago"].head()
## traz a coluna 2019/Ago
```

```
usaveis.loc["12 Acre"]
## localiza a linha cujo o indice se chama Acre
```
