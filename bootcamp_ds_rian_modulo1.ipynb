```
import pandas as pd

##importando dados por ano/mes processamento

dados = pd.read_csv("https://raw.githubusercontent.com/RianBrug/bootcamp-datascience-alura/main/procedimento-hospitalares-sus-por-local-internacao-valor-total-por-ano-mes-processamento.csv", 
                    encoding="ISO-8859-1", skiprows = 3, sep=";", 
                    skipfooter=12, thousands=".", decimal=",")
dados.head()
```

```
dados[["2008/Ago", "2008/Set"]].head()
##fatiamento dos dados do nosso dataframe
## para realizar uma análise, necessário limpar os arquivos
```

```
colunas_usaveis = dados.mean().index.tolist()
## mean ignora as colunas vazias. desta forma, os resultados, possuem todos os valores
## eliminação das colunas vazias é de acordo com a estratégia que esta sendo usada. 
## em outras situações, poderia ser aplicado estratégias para lidar com esses campos vazios
colunas_usaveis.insert(0, "Unidade da Federação")
##inserindo o valor 'UF', necessario transformar o index (série imutável) para uma lista de python
colunas_usaveis
```

## Desafio04: mudar palete de cores com o matplotlib.

```
dados_uf_tratados.head()
```

pesquisar sobre Seaborn (melhor representatividade de cores)

```
import seaborn as sns
from matplotlib.colors import ListedColormap

dados_uf_tratados["Total"] = dados_uf_tratados.sum(axis=1)
#criado coluna total da soma
#correção do eixo, param 'axis'

top_5_estados = dados_uf_tratados.sort_values(by='Total', ascending=False)
top_5_estados = top_5_estados.drop('Total', axis=1).head()
sns.set(color_codes=True, rc={"lines.linewidth": 2.0})
ax = top_5_estados.T.plot(figsize=(10,8))
ax.yaxis.set_major_formatter(ticker.EngFormatter())
plt.ylabel("Valor Gasto por Unidade da Federação em Milhões")
plt.xlabel("Ano/Mês do Atendimento")
plt.title("Os 5 estados que mais utilizaram recursos hospitalares - SUS")
plt.legend(title='Unidade da Federação', bbox_to_anchor=(1.05, 1.01), loc='best')
plt.show()
```
